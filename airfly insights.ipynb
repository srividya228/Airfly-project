```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
file_path = "flights_sample_3m.csv"   # Change path if needed
df = pd.read_csv(file_path, low_memory=False)

# Preview data
print("Shape of dataset:", df.shape)
print("\nColumns:\n", df.columns)
display(df.head())
```

```python
# Basic Info

print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary Statistics:\n")
display(df.describe(include="all"))
```

```python
# Handle Missing Values

# Fill numeric NaNs with 0 (delays, taxi times etc.)
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(0)

# Fill categorical NaNs with "Unknown"
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
# Distribution of Delays

plt.figure(figsize=(10,5))
sns.histplot(df['ARR_DELAY'], bins=50, kde=True, color="blue")
plt.title("Arrival Delay Distribution")
plt.xlabel("Delay (minutes)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['DEP_DELAY'], bins=50, kde=True, color="red")
plt.title("Departure Delay Distribution")
plt.xlabel("Delay (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Cancellation Analysis

cancel_map = {"A":"Carrier", "B":"Weather", "C":"NAS", "D":"Security"}
if 'CANCELLATION_CODE' in df.columns:
    df['CANCELLATION_REASON'] = df['CANCELLATION_CODE'].map(cancel_map).fillna("Not Cancelled")
    df['CANCELLATION_REASON'].value_counts().plot(kind='bar', figsize=(8,5))
    plt.title("Cancellation Reasons")
    plt.ylabel("Count")
    plt.show()
```

```python

#  Diversions

if 'DIVERTED' in df.columns:
    diversion_rate = df['DIVERTED'].mean() * 100
    print(f"Diversion Rate: {diversion_rate:.2f}%")
```

```python
#  Efficiency: Scheduled vs Actual Times

plt.figure(figsize=(10,5))
df[['CRS_ELAPSED_TIME','ELAPSED_TIME','AIR_TIME']].mean().plot(kind='bar')
plt.title("Average Scheduled vs Actual vs Air Time")
plt.ylabel("Minutes")
plt.show()
```

```python
# Delay Causes Breakdown

delay_cols = ['DELAY_DU_CARRIER','DELAY_DU_WEATHER','DELAY_DU_NAS',
              'DELAY_DU_SECURITY','DELAY_DU_LATE_AIRCRAFT']
present_delay_cols = [col for col in delay_cols if col in df.columns]

if present_delay_cols:
    df[present_delay_cols].sum().plot(kind='bar', figsize=(10,5))
    plt.title("Delay Causes Breakdown (Total Minutes)")
    plt.ylabel("Total Minutes of Delay")
    plt.show()
```
