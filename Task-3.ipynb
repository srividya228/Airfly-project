```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv("/content/flights_sample_3m.csv")

print("\n--- Shape of dataset ---")
print(df.shape)

print("\n--- Data Types & Nulls ---")
print(df.info())

```

```python
# Basic Info

print("\n--- Descriptive Stats ---")
print(df.describe(include='all'))
display(df.describe(include="all"))
```

```python
# Handle Missing Values

print("\n--- Missing Values ---")
print(df.isnull().sum())

# Fill categorical NaNs with "Unknown"
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
# --- Derived Features ---

# Ensure FL_DATE is datetime
if 'FL_DATE' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')

    # Month, Day of Week
    df['MONTH'] = df['FL_DATE'].dt.month
    df['DAY_OF_WEEK'] = df['FL_DATE'].dt.day_name()

# Extract Hour from DEP_TIME if present
if 'DEP_TIME' in df.columns:
    # Convert hhmm to hour safely
    df['DEP_HOUR'] = (df['DEP_TIME'] // 100).astype('Int64')
else:
    df['DEP_HOUR'] = None

# Create Route if ORIGIN and DEST are available
if 'ORIGIN' in df.columns and 'DEST' in df.columns:
    df['ROUTE'] = df['ORIGIN'].astype(str) + " → " + df['DEST'].astype(str)
else:
    df['ROUTE'] = None

# Only print columns that exist
cols_to_show = [col for col in ['FL_DATE','MONTH','DAY_OF_WEEK','DEP_HOUR','ROUTE'] if col in df.columns]
print(df[cols_to_show].head())
```

```python
# --- Save Preprocessed Data ---
df.to_csv("flights_preprocessed.csv", index=False)
print("Preprocessed dataset saved as flights_preprocessed.csv")
```

```python
# --- Feature Dictionary ---
feature_dict = {
    "FL_DATE": "Flight date (datetime)",
    "MONTH": "Month of flight (1–12)",
    "DAY_OF_WEEK": "Day of week (Monday–Sunday)",
    "DEP_HOUR": "Scheduled departure hour (0–23)",
    "ROUTE": "Origin → Destination route",
    # Add more key variables here
}
for k,v in feature_dict.items():
    print(f"{k}: {v}")
```

```python
# Time-based Analysis (if date columns exist)
if 'FL_DATE' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
    flights_per_month = df.groupby(df['FL_DATE'].dt.to_period('M')).size()
    flights_per_month.plot(kind='bar', figsize=(10,4))
```

```python
# Pairplot (using sample to avoid memory issues)

sample_df = df.sample(5000, random_state=42)
sns.pairplot(sample_df[['ARR_DELAY','DEP_DELAY','AIR_TIME','DISTANCE']])
plt.show()
```

```python
# Load dataset
df = pd.read_csv("/content/flights_preprocessed.csv")

import matplotlib.pyplot as plt
import seaborn as sns

# --- Top Airlines ---
if 'OP_CARRIER' in df.columns:
    df['OP_CARRIER'].value_counts().head(10).plot(
        kind='bar',
        figsize=(8,4),
        color=sns.color_palette("Paired", 10)   # 10 different colors
    )
    plt.title("Top 10 Airlines by Flight Count")
    plt.xlabel("Airline")
    plt.ylabel("Flights")
    plt.show()

# --- Top Routes ---
if 'ROUTE' in df.columns:
    df['ROUTE'].value_counts().head(10).plot(
        kind='bar',
        figsize=(10,4),
        color=sns.color_palette("Set2", 10)   # Softer distinct colors
    )
    plt.title("Top 10 Routes")
    plt.xlabel("Route")
    plt.ylabel("Flights")
    plt.show()

# --- Busiest Months ---
if 'MONTH' in df.columns:
    df['MONTH'].value_counts().sort_index().plot(
        kind='bar',
        figsize=(8,4),
        color=sns.color_palette("viridis", df['MONTH'].nunique()) # Gradient effect
    )
    plt.title("Flights by Month")
    plt.xlabel("Month")
    plt.ylabel("Flights")
    plt.show()
```

```python
#  Flight Distribution by Day of Week & Hour
# --- Flights by Day of Week ---
if 'DAY_OF_WEEK' in df.columns:
    df['DAY_OF_WEEK'].value_counts().reindex([
        'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
    ]).plot(kind='bar', figsize=(8,4))
    plt.title("Flights by Day of Week")
    plt.show()

# --- Flights by Departure Hour ---
if 'DEP_HOUR' in df.columns and df['DEP_HOUR'].notna().any():
    df['DEP_HOUR'].value_counts().sort_index().plot(kind='bar', figsize=(10,4))
    plt.title("Flights by Departure Hour")
    plt.show()
```

```python
# --- Delay Cause Analysis (Safe) ---
delay_cols = ['CARRIER_DELAY','WEATHER_DELAY','NAS_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY']

for col in delay_cols:
    if col in df.columns:
        if df[col].notna().sum() > 0 and df[col].sum() > 0:
            plt.figure(figsize=(8,4))
            df[col].dropna().plot(kind='hist', bins=30, alpha=0.6)
            plt.title(f"Distribution of {col}")
            plt.xlabel("Minutes Delay")
            plt.ylabel("Count")
            plt.show()
        else:
            print(f"Column {col} has no meaningful values (all NaN or zero).")
    else:
        print(f" Column {col} not found in dataset.")
```

```python
# --- Average Delay by Airline (Safe) ---

# Try multiple possible column names for carrier and arrival delay
carrier_col = None
for c in ['OP_CARRIER','AIRLINE','CARRIER','UniqueCarrier']:
    if c in df.columns:
        carrier_col = c
        break

arr_delay_col = None
for c in ['ARR_DELAY','ARRIVAL_DELAY','ArrDelay']:
    if c in df.columns:
        arr_delay_col = c
        break

if carrier_col and arr_delay_col:
    avg_delay = df.groupby(carrier_col)[arr_delay_col].mean().dropna().sort_values(ascending=False).head(10)
    if not avg_delay.empty:
        avg_delay.plot(
            kind='bar',
            figsize=(8,4),
            color=sns.color_palette("coolwarm", len(avg_delay))  # gradient color palette
        )
        plt.title("Top 10 Airlines by Average Arrival Delay")
        plt.ylabel("Average Delay (minutes)")
        plt.xlabel("Airline")
        plt.show()
    else:
        print(f" Column {arr_delay_col} has no valid values to plot.")
else:
    print(" Could not find suitable columns for airline or arrival delay in dataset.")
```

```python
# --- Average Delay by Origin Airport ---
if 'ORIGIN' in df.columns and 'ARR_DELAY' in df.columns:
    avg_delay_airport = df.groupby('ORIGIN')['ARR_DELAY'].mean().sort_values(ascending=False).head(10)
    avg_delay_airport.plot(kind='bar', figsize=(10,4))
    plt.title("Top 10 Airports by Average Arrival Delay")
    plt.ylabel("Average Delay (minutes)")
    plt.show()
```

```python
#  Distance vs Arrival Delay
if 'DISTANCE' in df.columns and 'ARR_DELAY' in df.columns:
    # Create bins of distance (e.g., every 500 miles)
    df['DISTANCE_BIN'] = pd.cut(df['DISTANCE'], bins=range(0, int(df['DISTANCE'].max())+500, 500))
    avg_delay = df.groupby('DISTANCE_BIN')['ARR_DELAY'].mean().reset_index()

    plt.figure(figsize=(10,5))
    sns.lineplot(x=avg_delay['DISTANCE_BIN'].astype(str), y=avg_delay['ARR_DELAY'], marker="o")
    plt.title("Average Arrival Delay by Distance Group")
    plt.xlabel("Flight Distance (binned, miles)")
    plt.ylabel("Average Delay (minutes)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
#  Boxplot of Delays by Airline

if 'ARR_DELAY' in df.columns and 'OP_UNIQUE_CARRIER' in df.columns:
    plt.figure(figsize=(12,6))
    sns.boxplot(x='OP_UNIQUE_CARRIER', y='ARR_DELAY', data=df)
    plt.title("Distribution of Arrival Delays by Airline")
    plt.xlabel("Airline")
    plt.ylabel("Arrival Delay (minutes)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# --- Flights per Year ---
if 'FL_DATE' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')
    flights_per_year = df.groupby(df['FL_DATE'].dt.year).size()

    flights_per_year.plot(
        kind='bar',
        figsize=(8,4),
        color=sns.color_palette("crest", len(flights_per_year))  # gradient by year
    )
    plt.title("Flights per Year")
    plt.xlabel("Year")
    plt.ylabel("Number of Flights")
    plt.show()
```

```python
# --- Flights per Month (Across Years) ---
if 'FL_DATE' in df.columns:
    flights_per_month = df.groupby(df['FL_DATE'].dt.month).size()

    flights_per_month.plot(kind='bar', figsize=(8,4))
    plt.title("Flights per Month (Aggregated)")
    plt.xlabel("Month (1 = Jan ... 12 = Dec)")
    plt.ylabel("Number of Flights")
    plt.show()
```

```python
# --- Average Arrival Delay per Month ---
if 'FL_DATE' in df.columns and 'ARR_DELAY' in df.columns:
    avg_delay_month = df.groupby(df['FL_DATE'].dt.month)['ARR_DELAY'].mean()

    avg_delay_month.plot(kind='line', marker='o', figsize=(8,4))
    plt.title("Average Arrival Delay per Month")
    plt.xlabel("Month")
    plt.ylabel("Avg Delay (minutes)")
    plt.show()
```

```python
# --- Average Arrival Delay per Year ---
if 'FL_DATE' in df.columns and 'ARR_DELAY' in df.columns:
    avg_delay_year = df.groupby(df['FL_DATE'].dt.year)['ARR_DELAY'].mean()

    avg_delay_year.plot(kind='line', marker='o', figsize=(8,4))
    plt.title("Average Arrival Delay per Year")
    plt.xlabel("Year")
    plt.ylabel("Avg Delay (minutes)")
    plt.show()
```

```python
# --- Monthly Cancellation Trends ---
if {'CANCELLED','FL_DATE'}.issubset(df.columns):
    df['MONTH'] = df['FL_DATE'].dt.month
    monthly_cancels = df.groupby('MONTH')['CANCELLED'].sum()
    monthly_cancels.plot(kind='line', marker='o', figsize=(8,4), color="tomato")
    plt.title("Monthly Flight Cancellations")
    plt.xlabel("Month")
    plt.ylabel("Cancellations")
    plt.show()
```

```python
# --- Cancellation Types (Carrier, Weather, NAS, Security) ---
cancel_map = {'A': 'Carrier', 'B': 'Weather', 'C': 'NAS', 'D': 'Security'}

if 'CANCELLATION_CODE' in df.columns:
    df['CANCEL_REASON'] = df['CANCELLATION_CODE'].map(cancel_map)
    cancel_counts = df['CANCEL_REASON'].value_counts()
    cancel_counts.plot(kind='bar', figsize=(6,4), color="orange")
    plt.title("Flight Cancellations by Reason")
    plt.xlabel("Reason")
    plt.ylabel("Count")
    plt.show()
```

```python
# --- Average Delay by Season ---
if 'FL_DATE' in df.columns and 'ARR_DELAY' in df.columns:
    df['MONTH'] = df['FL_DATE'].dt.month
    season_map = {12:'Winter',1:'Winter',2:'Winter',
                  3:'Spring',4:'Spring',5:'Spring',
                  6:'Summer',7:'Summer',8:'Summer',
                  9:'Fall',10:'Fall',11:'Fall'}
    df['SEASON'] = df['MONTH'].map(season_map)

    season_delay = df.groupby('SEASON')['ARR_DELAY'].mean().reindex(['Winter','Spring','Summer','Fall'])
    season_delay.plot(kind='bar', figsize=(6,4), color="teal")
    plt.title("Average Arrival Delay by Season")
    plt.ylabel("Avg Delay (minutes)")
    plt.show()
```

```python
# --- Cancellation Reasons ---
cancel_map = {'A':'Carrier', 'B':'Weather', 'C':'NAS', 'D':'Security'}

if 'CANCELLATION_CODE' in df.columns:
    df['CANCEL_REASON'] = df['CANCELLATION_CODE'].map(cancel_map)
    cancel_counts = df['CANCEL_REASON'].value_counts()
    cancel_counts.plot(kind='bar', figsize=(6,4), color="orange")
    plt.title("Cancellations by Reason")
    plt.xlabel("Reason")
    plt.ylabel("Count")
    plt.show()
```

```python
# --- Average Arrival Delay by Route Length (Paired Palette) ---
if {'DISTANCE','ARR_DELAY'}.issubset(df.columns):
    bins = [0, 500, 1500, 3000, df['DISTANCE'].max()]
    labels = ['Short (<500mi)','Medium (500-1500mi)','Long (1500-3000mi)','Ultra (>3000mi)']
    df['ROUTE_LENGTH'] = pd.cut(df['DISTANCE'], bins=bins, labels=labels)

    plt.figure(figsize=(8,5))
    sns.barplot(x='ROUTE_LENGTH', y='ARR_DELAY', data=df, estimator='mean', palette="Paired")
    plt.title("Average Arrival Delay by Route Length")
    plt.xlabel("Route Length")
    plt.ylabel("Average Delay (minutes)")
    plt.show()
```

```python
# --- Top 15 Airports by Flight Volume ---
if {'ORIGIN','ARR_DELAY'}.issubset(df.columns):
    airport_stats = df['ORIGIN'].value_counts().head(15)

    plt.figure(figsize=(12,6))
    airport_stats.plot(kind='barh', color=sns.color_palette("crest", 15))
    plt.title("Top 15 Origin Airports by Flight Count")
    plt.xlabel("Number of Flights")
    plt.ylabel("Airport")
    plt.show()
```

```python
# --- Cancellation Reasons by Month ---
if {'CANCELLATION_CODE','FL_DATE'}.issubset(df.columns):
    cancel_map = {'A':'Carrier','B':'Weather','C':'NAS','D':'Security'}
    df['CANCEL_REASON'] = df['CANCELLATION_CODE'].map(cancel_map)
    df['MONTH'] = df['FL_DATE'].dt.month

    # Create a flag for cancelled flights
    df['CANCEL_FLAG'] = df['CANCELLATION_CODE'].notna().astype(int)

    # Pivot: count cancellations by reason & month
    pivot_cancel = df.pivot_table(index='MONTH',
                                  columns='CANCEL_REASON',
                                  values='CANCEL_FLAG',
                                  aggfunc='sum').fillna(0)

    plt.figure(figsize=(10,6))
    sns.heatmap(pivot_cancel, cmap="Blues", annot=True, fmt=".0f")
    plt.title("Cancellations by Month & Reason")
    plt.show()
```
