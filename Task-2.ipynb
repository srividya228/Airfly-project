```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv("flights_sample_3m.csv")

print("\n--- Shape of dataset ---")
print(df.shape)

print("\n--- Data Types & Nulls ---")
print(df.info())

```

```python
# Basic Info

print("\n--- Descriptive Stats ---")
print(df.describe(include='all'))
display(df.describe(include="all"))
```

```python
# Handle Missing Values

print("\n--- Missing Values ---")
print(df.isnull().sum())

# Fill categorical NaNs with "Unknown"
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
# Univariate Analysis

# Numerical distributions
num_cols = df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"Distribution of {col}")plt.show()
```

```python
# Categorical distributions
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
plt.figure(figsize=(6,4))
df[col].value_counts().head(20).plot(kind='bar')
plt.title(f"Top 20 categories of {col}")
plt.show()
```

```python
# Bivariate Analysis

# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Time-based Analysis (if date columns exist)
if 'FL_DATE' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
    flights_per_month = df.groupby(df['FL_DATE'].dt.to_period('M')).size()
    flights_per_month.plot(kind='bar', figsize=(10,4))
```

```python
#  Target Variable Distribution (if predicting delays)
sns.histplot(df['ARR_DELAY'], bins=50, kde=True)
```

```python
# Pairplot (using sample to avoid memory issues)

sample_df = df.sample(5000, random_state=42)
sns.pairplot(sample_df[['ARR_DELAY','DEP_DELAY','AIR_TIME','DISTANCE']])
plt.show()
```

```python
#  Flights by Airline

if 'OP_UNIQUE_CARRIER' in df.columns:
   plt.figure(figsize=(10,5))
   df['OP_UNIQUE_CARRIER'].value_counts().plot(kind='bar')
   plt.title("Number of Flights by Airline")
   plt.xlabel("Airline")
   plt.ylabel("Flights Count")
   plt.show()
```

```python
#  Average Delay by Airline

if 'ARR_DELAY' in df.columns and 'OP_UNIQUE_CARRIER' in df.columns:
    plt.figure(figsize=(10,5))
    sns.barplot(x='OP_UNIQUE_CARRIER', y='ARR_DELAY', data=df)
    plt.title("Average Arrival Delay by Airline")
    plt.xlabel("Airline")
    plt.ylabel("Average Delay (minutes)")
    plt.show()
```

```python
#  Departure vs Arrival Delay (scatter)
if 'DEP_DELAY' in df.columns and 'ARR_DELAY' in df.columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(x='DEP_DELAY', y='ARR_DELAY', data=df, fill=True, cmap="mako", thresh=0.05)
    plt.title("2D Density: Departure vs Arrival Delay")
    plt.xlabel("Departure Delay (minutes)")
    plt.ylabel("Arrival Delay (minutes)")
    plt.show()
```

```python
#  Distance vs Arrival Delay
if 'DISTANCE' in df.columns and 'ARR_DELAY' in df.columns:
    # Create bins of distance (e.g., every 500 miles)
    df['DISTANCE_BIN'] = pd.cut(df['DISTANCE'], bins=range(0, int(df['DISTANCE'].max())+500, 500))
    avg_delay = df.groupby('DISTANCE_BIN')['ARR_DELAY'].mean().reset_index()

    plt.figure(figsize=(10,5))
    sns.lineplot(x=avg_delay['DISTANCE_BIN'].astype(str), y=avg_delay['ARR_DELAY'], marker="o")
    plt.title("Average Arrival Delay by Distance Group")
    plt.xlabel("Flight Distance (binned, miles)")
    plt.ylabel("Average Delay (minutes)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

if len(num_cols) > 1:
    sns.pairplot(df[num_cols].dropna(), diag_kind="kde")
    plt.suptitle("Pairplot of Numeric Features", y=1.02)
    plt.show()
```

```python
#  Boxplot of Delays by Airline
# --------------------
if 'ARR_DELAY' in df.columns and 'OP_UNIQUE_CARRIER' in df.columns:
    plt.figure(figsize=(12,6))
    sns.boxplot(x='OP_UNIQUE_CARRIER', y='ARR_DELAY', data=df)
    plt.title("Distribution of Arrival Delays by Airline")
    plt.xlabel("Airline")
    plt.ylabel("Arrival Delay (minutes)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
#  Flights per Month (time trend)
# --------------------
if 'FL_DATE' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')
    monthly_counts = df.groupby(df['FL_DATE'].dt.to_period('M')).size()
    monthly_counts.plot(kind='bar', figsize=(12,5))
    plt.title("Number of Flights per Month")
    plt.xlabel("Month")
    plt.ylabel("Flights Count")
    plt.show()
```

```python
#  Pairplot of Key Numerical Features

num_features = [col for col in ['ARR_DELAY','DEP_DELAY','DISTANCE'] if col in df.columns]
if len(num_features) > 1:
    sns.pairplot(df[num_features].dropna(), diag_kind='kde')
    plt.suptitle("Pairwise Relationships", y=1.02)
    plt.show()
```

```python
#  Average Arrival Delay by Day of Week

if 'FL_DATE' in df.columns and 'ARR_DELAY' in df.columns:
    df['FL_DATE'] = pd.to_datetime(df['FL_DATE'], errors='coerce')
    df['DAY_OF_WEEK'] = df['FL_DATE'].dt.day_name()
    plt.figure(figsize=(10,5))
    sns.barplot(x='DAY_OF_WEEK', y='ARR_DELAY', data=df,
                order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
    plt.title("Average Arrival Delay by Day of Week")
    plt.xlabel("Day of Week")
    plt.ylabel("Average Delay (minutes)")
    plt.show()
```

```python
# Cancellation Reasons

if 'CANCELLED' in df.columns and 'CANCELLATION_CODE' in df.columns:
    plt.figure(figsize=(6,5))
    df['CANCELLATION_CODE'].value_counts().plot(kind='bar', color='red')
    plt.title("Flight Cancellations by Reason")
    plt.xlabel("Reason Code")
    plt.ylabel("Number of Cancellations")
    plt.show()
```

```python
#  Delay Comparison: Origin vs Destination Airports

if 'ORIGIN' in df.columns and 'ARR_DELAY' in df.columns:
    top_airports = df['ORIGIN'].value_counts().head(5).index
    plt.figure(figsize=(10,6))
    sns.boxplot(x='ORIGIN', y='ARR_DELAY', data=df[df['ORIGIN'].isin(top_airports)])
    plt.title("Arrival Delays for Top 5 Origin Airports")
    plt.xlabel("Origin Airport")
    plt.ylabel("Arrival Delay (minutes)")
    plt.show()
```
